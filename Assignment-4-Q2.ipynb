```python
from numpy import asarray
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
```

# Problem 2) [3 pts] Application of Keras to build, compile, and train a neural network to perform XOR operation:

### a. Create an np array of shape 4x2 for the inputs and another 4x1 array for the labels of XOR

```python
features = np.array([[0,0], [0,1], [1,0], [1,1]])
labels = np.array([0,1,1,0])
classes=[0,1]

print('Input feature array shape: ', features.shape)
print('Input labels array shape: ', labels.shape)
```

### b. Plot the given data points with two different markers for each group.

```python
plt.plot(features[labels[:]==classes[0],0], features[labels[:]==classes[0],1], 'rs', 
         features[labels[:]==classes[1],0], features[labels[:]==classes[1],1], 'g^')
plt.axis([-0.5,1.5,-0.5,1.5])
plt.xlabel('x: feature 1')
plt.ylabel('y: feature 2')
plt.legend(['Class'+str(classes[0]), 'Class'+str(classes[1])])
plt.show()
```

### c. Based on the plot from part (b), what is the minimum number of layers and nodes that is required to classify the training data points correctly? Explain.

##### Answer: 
Based on the plot above, we will need 2 lines to separate the classes. Therefore we will need 2 nodes in first layer and 1 node in output layer.

### d. Build the network that you proposed in part c using the Keras library.

```python
model = keras.Sequential()

model.add(keras.layers.Dense(input_dim=2, units=2, activation='tanh'))
model.add(keras.layers.Dense(units=1, activation='sigmoid'))

model.summary()
```

### e. Compile the network. Make sure to select a correct loss function for this classification problem. Use stochastic gradient descent learning (SGD, learning rate of 0.1). Explain your selection of the loss function.

```python
model.compile(loss='binary_crossentropy', optimizer=tf.keras.optimizers.SGD(learning_rate=0.1), metrics=['accuracy'])
```

##### Answer:
For this classification, I chose 'binary_crossentropy' for loss function. The dataset has two class lables 1 or 0.
Cross-entropy calculates a score that summarizes the average difference between the actual and predicted probability distributions for predicting class 1 or 0.

### f. Train the network for 200 epochs and a batch size of 1.

```python
history=model.fit(features, labels, epochs=200, batch_size=1, verbose=1)
```

### g. Use the trained weights and plot the final classifier lines in the plot of part (b).

```python
def plot_classifier(features, labels, classes, weights):
    plt.plot(features[labels[:]==classes[0],0], features[labels[:]==classes[0],1], 'rs', 
         features[labels[:]==classes[1],0], features[labels[:]==classes[1],1], 'g^')
    plt.axis([-0.5,1.5,-0.5,1.5])
  
    for node_i in range(weights[0].shape[1]):
        thre_parms=np.array(weights[0][:,node_i])
        thre_parms=np.append(thre_parms, weights[1][node_i]) 
 
        x1 = np.linspace(-1,2,50)
        x2 = -(thre_parms[0]*x1+thre_parms[2])/thre_parms[1] 
        plt.plot(x1, x2, '-r')

    plt.xlabel('x: feature 1')
    plt.ylabel('y: feature 2')
    plt.legend(['Class'+str(classes[0]), 'Class'+str(classes[1])])
```

```python
weights=model.layers[0].get_weights() 
plot_classifier(features, labels, classes, weights)
```

### h. Plot the training loss (i.e., the learning curve) for all the epochs.

```python
def plot_curve(accuracy_train, loss_train):
  epochs=np.arange(loss_train.shape[0])
  plt.subplot(1,2,1)
  plt.plot(epochs,accuracy_train)
  #plt.axis([-1,2,-1,2])
  plt.xlabel('Epoch#')
  plt.ylabel('Accuracy')
  plt.title('Training Accuracy')

  plt.subplot(1,2,2)
  plt.plot(epochs,loss_train)
  plt.xlabel('Epoch#')
  plt.ylabel('Binary crossentropy loss')
  plt.title('Training loss')

  plt.show()
```

```python
acc_curve=np.array(history.history['accuracy'])
loss_curve=np.array(history.history['loss'])
plot_curve(acc_curve,loss_curve)
```
