Given two functions f1(x,y)=(x-2)^2 + (y-3)^2, and f2(x,y)=(1 - (y-3))^2 + 20((x+3) - (y-3)2)2, please implement gradient descent learning to search for global minimum value for each function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Use two Plots to show f1(x,y) and f2(x,y) values with respect to the two dimensional input x and y (You can specify the range of the x and y values)

```python
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
X, Y = np.meshgrid(x, y)
F1 = (X-2)**2 + (Y-3)**2

fig = plt.figure(figsize=(12,6))

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot_surface(X, Y, F1)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
ax1.set_title('f1')

F2 = (1 - (Y-3))**2 + 20*((X+3) - (Y-3)**2)**2
ax2 = fig.add_subplot(1,2,2, projection='3d')
ax2.plot_surface(X, Y, F2)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')
ax2.set_title('f2')

plt.show()
```

Starting from initial value: (x,y)=(0,0), use learning rate =0.5, report f1(x,y) and f2(x,y) values in T=10 iterations. 
(your code can report f1(x,y) and f2(x,y) values as tables, or simple print out the values)

```python
def GradientDescent_f1(eta, T):
    dash = '-' * 100

    print(dash)
    print('f1 =(x-2)**2 + (y-3)**2')
    print(dash)
    print('count\tx\t\ty\t\tdx\t\tdy\t\tx next\t\ty next')
    print(dash)

    x,y=0,0
    xx,yy=[], []
    for i in range(T):
        xx.append(x)
        yy.append(y)
    
        dfx = 2*x - 4
        dfy = 2*y - 6
    
        # next x,y
        new_x = x + eta * (-1) * dfx
        new_y = y + eta * (-1) * dfy
    
        print('%d\t%f\t%f\t%f\t%f\t%f\t%f' % (i, x, y, dfx, dfy, new_x, new_y))
        x= new_x
        y= new_y

    print(dash)
    
    plt.scatter(xx, yy)
    plt.show()
```

```python
#f1 =(x-2)**2 + (y-3)**2

eta=0.5
T = 10
GradientDescent_f1(eta, T)
```

```python
def GradientDescent_f2(eta, T):
    dash = '-' * 100

    print(dash)
    print('f2 = (1 - (y-3))**2 + 20*((x+3) - (y-3)**2)**2')
    print(dash)
    print('count\tx\t\ty\t\tdx\t\tdy\t\tx next\t\ty next')
    print(dash)

    x,y=0,0
    xx,yy=[], []
    
    try:
        for i in range(T):
            xx.append(x)
            yy.append(y)
    
            dfx = 40*x - 40*(y - 3)**2 + 120
            dfy = 2*y + 20*(12 - 4*y)*(x - (y - 3)**2 + 3) - 8
    
            # next x,y
            new_x = x + eta * (-1) * dfx
            new_y = y + eta * (-1) * dfy
    
            print('%d\t%f\t%f\t%f\t%f\t%f\t%f' % (i, x, y, dfx, dfy, new_x, new_y))
            x= new_x
            y= new_y
        
    except OverflowError as err:
        print ('Overflowed at iteration: ', i)

    print(dash)
    
    plt.scatter(xx, yy)
    plt.show()
```

```python
f2 = (1 - (y-3))**2 + 20*((x+3) - (y-3)**2)**2

eta=0.5
T = 10
GradientDescent_f2(eta, T)
```
